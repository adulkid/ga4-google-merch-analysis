```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
sns.set_theme(style="whitegrid", font='AppleGothic')

df = pd.read_csv('explore_depth.csv')  # fail_type, depth_bin, session_cnt, pct

label_map = {'explore_fail': '탐색 후 미도달', 'pass_to_view_item': '도달'}
df = df[df['fail_type'].isin(label_map)].copy()
df['그룹'] = df['fail_type'].map(label_map)

# depth_bin 정렬 (0이 있을 수 있으니 데이터 보고 조정)
bin_order = ['0', '1', '2', '3', '4', '5+']
bin_order = [b for b in bin_order if b in df['depth_bin'].values]

fig, ax = plt.subplots(figsize=(10, 5.5))
sns.barplot(
    data=df, x='depth_bin', y='pct', hue='그룹',
    order=bin_order,
    hue_order=['탐색 후 미도달', '도달'],
    palette={'탐색 후 미도달': '#3D5AFE', '도달': '#C9CDD2'},
    ax=ax,
)
ax.set_xlabel('세션 내 고유 페이지 조회 수', fontsize=11)
ax.set_ylabel('세션 비율 (%)', fontsize=11)
ax.set_title('탐색 깊이 분포',
             fontsize=14, fontweight='bold', loc='left', pad=15)
ax.legend(title='', fontsize=11)
for c in ax.containers:
    ax.bar_label(c, fmt='%.0f', fontsize=8, padding=2, color='#666666')
sns.despine()
plt.tight_layout()
plt.savefig('03_explore_depth.png', dpi=200, bbox_inches='tight')
plt.show()
```
