```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df2 = pd.read_csv('02_channel_conversion.csv')
df2_sorted = df2.sort_values('CVR', ascending=False)
print(df2.shape)
df2_sorted.head()
```

```python
channel_colors = {
    'organic': '#4CAF50', 
    'direct': '#2196F3', 
    'referral': '#FF9800', 
    'unknown': '#9E9E9E', 
    'paid': '#F44336'
}

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('채널별 성과 분석', fontsize=16, fontweight='bold')

# subplot 1 - 세션 수 + CVR 이중축
ax1 = axes[0]
ax2 = ax1.twinx()
ax1.bar(
    df2_sorted['channel'], 
    df2_sorted['total_session_cnt'], 
    color=[channel_colors[c] for c in df2_sorted['channel']]
)
ax2.plot(
    df2_sorted['channel'], 
    df2_sorted['CVR'], 
    color='black', 
    marker='o', 
    linewidth=1, 
    markersize=4
)
# 바차트 수치 - 막대 위에
for i, (channel, value) in enumerate(zip(df2_sorted['channel'], df2_sorted['total_session_cnt'])):
    ax1.text(i, value + 1000, f'{value:,}', ha='center', fontsize=8)
# CVR 수치 - 점 옆에
for i, (channel, value) in enumerate(zip(df2_sorted['channel'], df2_sorted['CVR'])):
    ax2.text(i + 0.1, value, f'{value:.2f}%', ha='left', fontsize=8)
ax1.set_xlabel('채널')
ax1.set_ylabel('세션 수', color='black')
ax2.set_ylabel('CVR (%)', color='black')
axes[0].set_title('채널별 세션 수 & CVR')

x = np.arange(len(df2_sorted))
width = 0.35

# avg 막대 - 왼쪽
axes[1].bar(x - width/2, df2_sorted['revenue_per_session_avg'], 
            width=width, label='평균', color='steelblue')

# med 막대 - 오른쪽
axes[1].bar(x + width/2, df2_sorted['revenue_per_session_med'], 
            width=width, label='중앙값', color='coral')

# revenue (avg, med) 수치
for i, (avg, med) in enumerate(zip(df2_sorted['revenue_per_session_avg'], 
                                    df2_sorted['revenue_per_session_med'])):
    axes[1].text(i - width/2, avg + 0.5, f'$ {avg:.1f}', ha='center', fontsize=8)
    axes[1].text(i + width/2, med + 0.5, f'$ {med:.1f}', ha='center', fontsize=8)

# x축 채널명으로 변경
axes[1].set_xticks(x)
axes[1].set_xticklabels(df2_sorted['channel'])
axes[1].set_title('채널별 세션당 매출')
axes[1].set_ylabel('매출 (USD)')
axes[1].legend()
plt.savefig('채널별 성과 분석.png', dpi=300, bbox_inches='tight')
```

```python
df3_total = pd.read_csv('03_funnel_total.csv')
df3_total_sorted = df3_total.sort_values('step', ascending=True)
base = df3_total_sorted[df3_total_sorted['stage'] == 'session_start']['cnt'].values[0]
df3_total_sorted['cumulative_rate'] = (df3_total_sorted['cnt'] / base * 100).round(1)
df3_total_sorted
```

```python
df3_total_sorted = df3_total_sorted.copy()
df3_total_sorted['prev_cnt'] = df3_total_sorted['cnt'].shift(1)
df3_total_sorted['retention_rate'] = (
    df3_total_sorted['cnt'] / df3_total_sorted['prev_cnt'] * 100
).round(1)
```

```python
x = list(range(len(df3_total_sorted)))
stages = df3_total_sorted['stage'].tolist()

fig, ax1 = plt.subplots(figsize=(12, 6))

# 왼쪽 Y축 — 누적 잔존율 (line)
ax1.plot(x, df3_total_sorted['cumulative_rate'],
         marker='o', linewidth=2, markersize=5,
         color='steelblue', label='누적 잔존율', zorder=5)
ax1.fill_between(x, df3_total_sorted['cumulative_rate'],
                 alpha=0.1, color='steelblue')
ax1.set_ylabel('누적 잔존율 (%)', color='steelblue')
ax1.set_ylim(0, 115)

for i, (rect, color) in enumerate(zip(ax2.patches, colors)):
    rect.set_alpha(0.6 if color == '#e74c3c' else 0.25)

for i, val in enumerate(df3_total_sorted['cumulative_rate']):
    is_viewitem = df3_total_sorted['stage'].iloc[i] == 'view_item'
    y_offset = -10 if is_viewitem else 3
    ax1.text(i, val + y_offset, f'{val:.1f}%', ha='center',
             fontsize=10,
             fontweight='bold',
             color='#e74c3c' if is_viewitem else 'black')

# 오른쪽 Y축 — 전 단계 대비 잔존율 (bar)
ax2 = ax1.twinx()
colors = ['#e74c3c' if i == 1 else '#95a5a6'
          for i in range(len(df3_total_sorted))]
ax2.bar(x, df3_total_sorted['retention_rate'],
        color=colors, alpha=0.3, width=0.4, label='전 단계 대비 잔존율')
ax2.set_ylabel('전 단계 대비 잔존율 (%)', color='gray')
ax2.set_ylim(0, 130)

for i, val in enumerate(df3_total_sorted['retention_rate']):
    is_viewitem = df3_total_sorted['stage'].iloc[i] == 'view_item'
    if pd.notna(val) and not is_viewitem:
        ax2.text(i, val + 2, f'{val:.1f}%', ha='center',
                 fontsize=8, color='gray')

ax1.set_xticks(x)
ax1.set_xticklabels(stages, rotation=15, ha='right')
ax1.set_title('전체 퍼널 누적 잔존율 & 전 단계 대비 잔존율', fontsize=12, fontweight='bold')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.tight_layout()
plt.savefig('funnel_dual_axis.png', dpi=300, bbox_inches='tight')
plt.show()
```
