```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# ── 데이터 로드 ──
df = pd.read_csv('first_funnel_type.csv')
order = ['즉시 이탈', '탐색 후 이탈', '도달']
df = df.set_index('session_type').loc[order].reset_index()

colors = {
    '즉시 이탈': '#C9CDD2',
    '탐색 후 이탈': '#3D5AFE',
    '도달': '#8A94A6',
}
color_list = [colors[t] for t in df['session_type']]

fig, ax = plt.subplots(figsize=(7, 7))

wedges, _ = ax.pie(
    df['percentage'],
    colors=color_list, 
    startangle=90,
    counterclock=False,
    wedgeprops=dict(width=0.42, edgecolor='white', linewidth=3),  # width<1 = 도넛
)

# ── 조각 바깥 라벨 (선 연결) ──
for w, (_, row) in zip(wedges, df.iterrows()):
    ang = (w.theta2 + w.theta1) / 2
    x = np.cos(np.radians(ang))
    y = np.sin(np.radians(ang))
    ha = 'left' if x > 0 else 'right'
    is_target = row['session_type'] == '탐색 후 이탈'
    ax.annotate(
        f"{row['session_type']}\n({row['total_sessions']:,})\n{row['percentage']}%",
        xy=(x*0.79, y*0.79),
        xytext=(x*1.15, y*1.15),
        ha=ha, va='center',
        fontsize=12,
        fontweight='bold' if is_target else 'normal',
        color='#3D5AFE' if is_target else '#555555',
        arrowprops=dict(arrowstyle='-', color='#BBBBBB', lw=1),
    )

# ── 도넛 가운데: 핵심 메시지 ──
target = df[df['session_type'] == '탐색 후 이탈'].iloc[0]
ax.text(0, 0.15, f"전체 세션 중", ha='center', va='center',
        fontsize=12, fontweight='bold')
ax.text(0, 0, f"{target['percentage']}%", ha='center', va='center',
        fontsize=34, fontweight='bold', color='#3D5AFE')
ax.text(0, -0.15, f"탐색 후 이탈하는 모습", ha='center', va='center',
        fontsize=12, fontweight='bold')

ax.set_title('첫 진입 세션의 행동 분류', fontsize=15, fontweight='bold', pad=20)
ax.axis('equal')

plt.tight_layout()
plt.savefig('01_session_classification.png', dpi=200, bbox_inches='tight')
plt.show()
```
