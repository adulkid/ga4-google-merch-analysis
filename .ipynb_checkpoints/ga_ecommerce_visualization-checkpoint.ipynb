```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

channel_colors = {
    'organic': '#4CAF50', 
    'direct': '#2196F3', 
    'referral': '#FF9800', 
    'unknown': '#9E9E9E', 
    'paid': '#F44336'
}

sns.set_style('ticks')

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df2 = pd.read_csv('02_channel_conversion.csv')
print(df2.shape)
df2.head()
```

```python
df2_sorted = df2.sort_values('CVR', ascending=False)
df2_sorted.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('채널별 성과 분석', fontsize=16, fontweight='bold')

# subplot 1 - 세션 수 + CVR 이중축
ax1 = axes[0]
ax2 = ax1.twinx()
ax1.bar(
    df2_sorted['channel'], 
    df2_sorted['total_session_cnt'], 
    color=[channel_colors[c] for c in df2_sorted['channel']]
)
ax2.plot(
    df2_sorted['channel'], 
    df2_sorted['CVR'], 
    color='black', 
    marker='o', 
    linewidth=1, 
    markersize=4
)
# 바차트 수치 - 막대 위에
for i, (channel, value) in enumerate(zip(df2_sorted['channel'], df2_sorted['total_session_cnt'])):
    ax1.text(i, value + 1000, f'{value:,}', ha='center', fontsize=8)
# CVR 수치 - 점 옆에
for i, (channel, value) in enumerate(zip(df2_sorted['channel'], df2_sorted['CVR'])):
    ax2.text(i + 0.1, value, f'{value:.2f}%', ha='left', fontsize=8)
ax1.set_xlabel('채널')
ax1.set_ylabel('세션 수', color='black')
ax2.set_ylabel('CVR (%)', color='black')
axes[0].set_title('채널별 세션 수 & CVR')

x = np.arange(len(df2_sorted))
width = 0.35

# avg 막대 - 왼쪽
axes[1].bar(x - width/2, df2_sorted['revenue_per_session_avg'], 
            width=width, label='평균', color='steelblue')

# med 막대 - 오른쪽
axes[1].bar(x + width/2, df2_sorted['revenue_per_session_med'], 
            width=width, label='중앙값', color='coral')

# revenue (avg, med) 수치
for i, (avg, med) in enumerate(zip(df2_sorted['revenue_per_session_avg'], 
                                    df2_sorted['revenue_per_session_med'])):
    axes[1].text(i - width/2, avg + 0.5, f'$ {avg:.1f}', ha='center', fontsize=8)
    axes[1].text(i + width/2, med + 0.5, f'$ {med:.1f}', ha='center', fontsize=8)

# x축 채널명으로 변경
axes[1].set_xticks(x)
axes[1].set_xticklabels(df2_sorted['channel'])
axes[1].set_title('채널별 세션당 매출')
axes[1].set_ylabel('매출 (USD)')
axes[1].legend()
plt.savefig('채널별 성과 분석.png', dpi=300, bbox_inches='tight')
```

## 채널별 유입 및 전환율 분석 요약
### - 브랜드 인지도 기반 유입(referral)이 구매 의도가 높은 유저를 데려오는 반면, 광고 유입(paid)은 탐색 목적 유저 비중이 높은 것으로 추정
### - unknown 채널은 전환율 CVR이 높게(1.78%) 나타나나, 쿠키 수집 미동의 유저로 분류되어 채널 특성 해석에 한계가 존재
### - 전 채널 공통으로 세션당 매출이 평균 값이 중앙 값보다 높게 나타나 고매출 세션이 평균을 왜곡하는 우편향 분포 확인. 채널 간 실질적 구매 가치의 차이는 크지 않음

```python
df3_total = pd.read_csv('03_funnel_total.csv')
df3_channel = pd.read_csv('03_funnel_channel.csv')
```

```python
print(df3_total.shape)
df3_total.head(10)
```

```python
print(df3_channel.shape)
df3_channel
```

```python
df3_total_sorted = df3_total.sort_values('step', ascending=True)
base = df3_total_sorted[df3_total_sorted['stage'] == 'session_start']['cnt'].values[0]
df3_total_sorted['cumulative_rate'] = (df3_total_sorted['cnt'] / base * 100).round(1)
```

```python
x = list(range(len(df3_total_sorted)))

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, df3_total_sorted['cumulative_rate'],
    marker='o', linewidth=1, markersize=3, color='steelblue')

ax.fill_between(x, df3_total_sorted['cumulative_rate'],
    alpha=0.15, color='steelblue')

for i, rate in enumerate(df3_total_sorted['cumulative_rate']):
    ax.text(i, rate + 3, f'{rate:.1f}%', ha='center', fontsize=9, fontweight='bold')

ax.set_xticks(x)
ax.set_xticklabels(df3_total_sorted['stage'], rotation=15, ha='right')
ax.set_title('전체 퍼널 누적 잔존율')
ax.set_ylabel('잔존율 (%)')
ax.set_ylim(0, 115)
plt.tight_layout()
plt.savefig('전체 퍼널 누적 잔존율.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
base = df3_channel[df3_channel['stage'] == 'session_start'][['channel', 'cnt']].rename(columns={'cnt': 'base_cnt'})
df3_channel = df3_channel.merge(base, on='channel')
df3_channel['cumulative_rate'] = (df3_channel['cnt'] / df3_channel['base_cnt'] * 100).round(1)
```

```python
df3_channel
```

```python
# 전체 퍼널 기준으로 보여주니 차이가 명확히 보이지 않아서 첫 퍼널 제외
df3_channel_filtered = df3_channel[df3_channel['step'] >= 2]
df3_channel_filtered['cumulative_rate'].max()
```

```python
markers = {'direct': 'D', 'organic': 'o', 'paid': '^', 'referral': 's', 'unknown': 'x'}
fig, ax = plt.subplots(figsize=(12, 6))

for channel, group in df3_channel_filtered.groupby('channel'):
    group_sorted = group.sort_values('step')
    ax.plot(
        group_sorted['stage'], 
        group_sorted['cumulative_rate'], 
        marker = markers[channel], 
        color=channel_colors[channel], 
        linewidth=1.5, 
        markersize=3, 
        label=channel
    )
# 각 stage별 max/min 채널 찾기
for stage, group in df3_channel_filtered.groupby('stage'):
    max_row = group.loc[group['cumulative_rate'].idxmax()]
    min_row = group.loc[group['cumulative_rate'].idxmin()]
    
    x_pos = list(df3_channel_filtered['stage'].unique()).index(stage)
    
    # 최대값
    ax.text(x_pos, max_row['cumulative_rate'] + 0.3, 
            f"{max_row['cumulative_rate']}%", 
            ha='center', fontsize=8, color=channel_colors[max_row['channel']])
    # 최소값
    ax.text(x_pos, min_row['cumulative_rate'] - 0.8, 
            f"{min_row['cumulative_rate']}%", 
            ha='center', fontsize=8, color=channel_colors[min_row['channel']])
ax.set_ylim(0, 25)
ax.legend()
plt.xticks(rotation=15)
ax.set_title('채널별 퍼널 잔존율 (session_start 기준 누적, view_item 이후)')
ax.set_ylabel('잔존율 (%)')
plt.tight_layout()
plt.savefig('채널별 퍼널 잔존율.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 채널별 퍼널 잔존율 분석 요약
### - 전체 세션 78%가 session_start -> view_item 구간에서 이탈. 상품 상세 페이지 진입 자체가 낮아 랜딩 페이지 또는 상품 노출 방식 개선이 필요할 수 있음
### - add_to_cart -> begin_checkout 구간에서 두번째 주요 이탈 발생. 장바구니 이후 결제 진입 과정에서 UX 마찰이 존재할 가능성 있으며 전체 결제 플로우에 대한 점검 필요
### - add_shipping_info -> add_payment_info 구간은 전 퍼널 단계 대비 (61% 잔존) 이탈율이 낮지 않음

```python
df4 = pd.read_csv('04_user_engagement.csv')
print(df4.shape)
df4.head(15)
```

```python
df4_group = df4.groupby('user_group').agg(
    page_view_med=('page_view_med', 'mean'),
    engagement_min_med=('engagement_min_med', 'mean')
).reset_index()
```

```python
group_colors = {
    '비구매자': '#d9534f',        # 빨강 - 전환 안 됨
    '구매자_비구매세션': '#f0ad4e', # 주황 - 잠재 구매자
    '구매자_구매세션': '#5cb85c'    # 초록 - 전환 완료
}
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 그룹 순서 정렬 (비구매자 → 구매자_비구매 → 구매자_구매)
order = ['비구매자', '구매자_비구매세션', '구매자_구매세션']
df4_group = df4_group.set_index('user_group').loc[order].reset_index()
colors = [group_colors[g] for g in df4_group['user_group']]

# subplot 1 - page_view_med
axes[0].bar(df4_group['user_group'], df4_group['page_view_med'], color=colors)
axes[0].set_title('평균 페이지 뷰 수')

# subplot 2 - engagement_min_med
axes[1].bar(df4_group['user_group'], df4_group['engagement_min_med'], color=colors)
axes[1].set_title('평균 체류 시간(분)')

for i, value in enumerate(df4_group['page_view_med']):
    axes[0].text(i, value, f'{value:.1f}개', ha='center', fontsize=10, fontweight='bold', 
    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

for i, value in enumerate(df4_group['engagement_min_med']):
    axes[1].text(i, value, f'{value:.1f}분', ha='center', fontsize=10, fontweight='bold', 
    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))
    
fig.suptitle('구매자_구매세션은 비구매자 대비 페이지뷰 12배, 체류시간 63배 [세션 단위 평균]', 
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('구매자 - 비구매자 세션 단위 분석.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 구매자와 비구매자의 세션 단위 분석 요약
### - 구매자의 비구매 세션에서도 비구매 유저의 세션보다 페이지뷰(1.7배), 체류시간(4배) 가량 높게 나타났다
### - 구매자는 비구매 세션에서도 탐색 깊이가 다르다는 것을 보여주며 리타겟팅 대상으로서 가치가 있음을 보여준다

```python
df5 = pd.read_csv('05_retention_week.csv')
df5
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

df5_filtered = df5[df5['week_num'] > 0]

ax.plot(df5_filtered['week_num'], df5_filtered['retention_rate'],
        marker='o', linewidth=1, markersize=4, color='steelblue')

ax.fill_between(df5_filtered['week_num'], df5_filtered['retention_rate'],
                alpha=0.15, color='steelblue')

for _, row in df5_filtered.iterrows():
    ax.text(row['week_num'], row['retention_rate'] + 0.05,
            f"{row['retention_rate']}%",
            ha='center', fontsize=8, fontweight='bold')

ax.annotate('[W0] 95.18% → [W1] 1.84%\n첫 주 이후 급격한 이탈 발생',
            xy=(1, 1.84), xytext=(4, 1.2),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=8, color='red')

ax.set_title('주차별 리텐션율 (W1~W12)', fontsize=13, fontweight='bold')
ax.set_xlabel('Week')
ax.set_ylabel('리텐션율 (%)')
ax.set_xticks(df5_filtered['week_num'])
ax.set_xticklabels([f'W{int(w)}' for w in df5_filtered['week_num']])
plt.tight_layout()
plt.savefig('주차별 리텐션율.png', dpi=300, bbox_inches='tight')
plt.show()
```
