```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
sns.set_theme(style="whitegrid", font='AppleGothic')

# ── 데이터 로드 ──
df = pd.read_csv('category_demand.csv')  # 컬럼: fail_type, category, ratio

# instantly_fail 제외 (카테고리 거의 안 봄 - 비교 무의미)
df = df[df['fail_type'].isin(['explore_fail', 'pass_to_view_item'])].copy()

# 비율 → %
df['pct'] = df['ratio'] * 100

# 라벨 한글화
label_map = {'explore_fail': '탐색 후 미도달', 'pass_to_view_item': '도달'}
df['그룹'] = df['fail_type'].map(label_map)

# 카테고리 정렬: 도달 기준 비율 높은 순
cat_order = (df[df['fail_type']=='pass_to_view_item']
             .sort_values('pct', ascending=False)['category'].tolist())

fig, ax = plt.subplots(figsize=(13, 6))

sns.barplot(
    data=df, x='category', y='pct', hue='그룹',
    order=cat_order,
    hue_order=['탐색 후 미도달', '도달'],
    palette={'탐색 후 미도달': '#3D5AFE', '도달': '#C9CDD2'},
    ax=ax,
)

ax.set_xlabel('')
ax.set_ylabel('세션 중 카테고리 도달 비율 (%)', fontsize=11)
ax.set_title('그룹별 카테고리 수요 비교',
             fontsize=15, fontweight='bold', loc='left', pad=15)
plt.xticks(rotation=35, ha='right', fontsize=10)
ax.legend(title='', fontsize=11, loc='upper right')

# 막대 위 값 라벨
for c in ax.containers:
    ax.bar_label(c, fmt='%.1f', fontsize=8, padding=2, color='#666666')

sns.despine()
plt.tight_layout()
plt.savefig('02_category_demand.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
# ── 데이터 로드 ──
df = pd.read_csv('category_demand.csv')  # 컬럼: fail_type, category, ratio
df['pct'] = df['ratio'] * 100

# 라벨 한글화
label_map = {
    'explore_fail': '탐색 후 미도달',
    'pass_to_view_item': '도달',
    'instantly_fail': '즉시 이탈',
}
df['그룹'] = df['fail_type'].map(label_map)

# ── 상위 8개 카테고리: 도달 기준 비율 높은 순 ──
top8 = (df[df['fail_type'] == 'pass_to_view_item']
        .sort_values('pct', ascending=False)
        .head(8)['category'].tolist())
df = df[df['category'].isin(top8)]

fig, ax = plt.subplots(figsize=(14, 6))

sns.barplot(
    data=df, x='category', y='pct', hue='그룹',
    order=top8,
    hue_order=['탐색 후 미도달', '도달', '즉시 이탈'],
    palette={
        '탐색 후 미도달': '#3D5AFE',   # 타겟 강조
        '도달': '#C9CDD2',            # 회색
        '즉시 이탈': '#FF9F1C',        # 구분용 (대조)
    },
    ax=ax,
)

ax.set_xlabel('')
ax.set_ylabel('세션 중 카테고리 도달 비율 (%)', fontsize=11)
ax.set_title('그룹별 카테고리 수요 비교 (상위 8개)',
             fontsize=15, fontweight='bold', loc='left', pad=15)
plt.xticks(rotation=30, ha='right', fontsize=10)
ax.legend(title='', fontsize=11, loc='upper right')

for c in ax.containers:
    ax.bar_label(c, fmt='%.1f', fontsize=7, padding=2, color='#666666')

sns.despine()
plt.tight_layout()
# plt.savefig('02_category_demand_3group.png', dpi=200, bbox_inches='tight')
plt.show()
```
